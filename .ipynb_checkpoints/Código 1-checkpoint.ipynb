```python
#Importação das bibliotecas
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Variável para inicializar o navegador
driver = webdriver.Edge()  # Para utilizar essa função é necessário ter o msedgedriver instalado e no PATH, segue link para download "https://developer.microsoft.com/pt-br/microsoft-edge/tools/webdriver?utm_source=chatgpt.com&form=MA13LH#downloads"
```

```python
# Variável que permite esperar o carregamento da página antes de rodar o script
wait = WebDriverWait(driver, 20)
```

```python
# Navegação a página web em que serão extraídos os dados
driver.get("https://www3.bcb.gov.br/ifdata/")  # Substitua pela URL real
```

```python
# Configuração de navegação na página, onde se faz necessário interação com alguns botões para se chegar na tabela em questão
wait.until(EC.element_to_be_clickable((By.ID, "btnDataBase"))).click()
wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'12/2024')]"))).click()

wait.until(EC.element_to_be_clickable((By.ID, "btnTipoInst"))).click()
wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'Conglomerados Financeiros e Instituições Independentes')]"))).click()

wait.until(EC.element_to_be_clickable((By.ID, "btnRelatorio"))).click()
wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'Resumo')]"))).click()
```

```python
# Função para esperar a tabela que buscamos ficar visível
wait.until(EC.visibility_of_element_located((By.ID, "dataTable")))
```

```python
# Captura o HTML da tabela
html_tabela = driver.find_element(By.ID, "dataTable").get_attribute("outerHTML")
```

```python
# Como a tabela foi inserida na página através de um html é necessário utilizar o BeautifulSoup para fazer parsing
soup = BeautifulSoup(html_tabela, "html.parser")
```

```python
# Localizar a tabela
table = soup.find("table", {"id": "dataTable"})
```

```python
# Extrair o cabeçalho da tabela
headers = [th.text.strip() for th in table.find_all("th")]
```

```python
# Extrair todas as linhas da tabela
rows = []
for tr in table.find_all("tr"):
    cells = tr.find_all("td")
    if cells:
        rows.append([cell.text.strip() for cell in cells])
```

```python
# Converter os dados em HTML para um formato editável e posteriormente ter a possibilidade de exportar
df = pd.DataFrame(rows, columns=headers)
display(df)

#Espaço para tratar a tabela caso haja necessidade
#Espaço para tratar a tabela caso haja necessidade

# Salvando em CSV
df.to_csv("tabela_bcb.csv", index=False, encoding='utf-8-sig')  # 'utf-8-sig' garante compatibilidade com Excel

#Função para fechar o navegador após o download do arquivo
driver.quit()
```
